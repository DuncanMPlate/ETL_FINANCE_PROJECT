```python
 import pandas as pd
from sqlalchemy import create_engine
```

```python
csv_file_1 ="../group_project_2/Resources/indexData.csv"
indexData_df= pd.read_csv(csv_file_1)
indexData_df.head()
```

```python
csv_file_2 ="../group_project_2/Resources/indexInfo.csv"
indexInfo_df= pd.read_csv(csv_file_2)
indexInfo_df.head()
```

```python
csv_file_3 ="../group_project_2/Resources/indexProcessed.csv"
indexProcessed_df= pd.read_csv(csv_file_3)
indexProcessed_df.head()
```

```python
rds_connection_string = "<username>:<password>@localhost:5432/stock_exchange"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
engine.table_names()
```

```python
indexInfo_df.to_sql(name='indexinfo', con=engine, if_exists='append', index=True, index_label='id')
```

```python
indexData_df.to_sql(name='indexdata', con=engine, if_exists='append', index=True, index_label='id')
```

```python
pd.read_sql_query('select * from indexinfo', con=engine).head()
```
